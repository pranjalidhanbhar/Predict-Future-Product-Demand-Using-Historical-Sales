# Feature Engineering

```python
df.columns
```

```python
## Final Selling Price
```

```python
df['Final_Price'] = df['Price'] - df['Discount_%']
```

```python
df['Final_Price']
```

```python
## Create Discount Amount
```

```python
df['Discount_%'] = (df['Price'] * df['Discount_%']) / 100
```

```python
df['Discount_%']
```

```python
## Delivery Time (in Days)
```

```python
df['Delivery_Time'] = (df['Delivery_Date'] - df['Order_Date']).dt.days
```

```python
df['Delivery_Time']
```

```python
## Category Profit Analysis
```

```python
df.groupby('Payment_Method')['Sales'].mean().sort_values(ascending=False)
```

```python
## High Sales but Low Profit
```

```python
products = df[(df['Sales'] > df['Sales'].mean()) & 
              (df['Profit'] < df['Profit'].mean())]

products[['Product_Name','Sales','Profit']]
```

```python
## Delivery Time Impact on Rating
```

```python
df['Delivery_Time'] = (df['Delivery_Date'] - df['Order_Date']).dt.days
df[['Delivery_Time','Rating']].corr()
```
